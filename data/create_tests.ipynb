```python
import pandas as pd

df = pd.read_csv("testdist.tsv", header=None, sep="\t")
df2 = pd.read_csv("dict.tsv", header=None, sep="\t")

df
```

```python
def get_english(feat):
    swahili_word = feat.split("+")[-1]
    # print(swahili_word)
    return df2[df2[0] == swahili_word].values[0][1]

def create_eng_feat(feat):
    return "+".join(feat.split("+")[:-1] )
```

```python
df[3] = df[0].apply(lambda x:create_eng_feat(x)+"+"+ get_english(x))
```
